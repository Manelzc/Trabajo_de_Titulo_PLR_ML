### Separación dataset

```python
def split_eye_data(dataset):
    # Filtrar datos para el ojo izquierdo y el ojo derecho
    left_eye_data = dataset[dataset['eye_id'] == 1]
    right_eye_data = dataset[dataset['eye_id'] == 0]

    return left_eye_data, right_eye_data
```

```python
# Llamar a la función para separar los datos
C001_L, C001_R = split_eye_data(C001_PP)

# Verificar los primeros registros de cada dataset
print("Datos del ojo izquierdo:")
display(C001_L.head(10))
print(C001_L.shape)

print("\nDatos del ojo derecho:")
display(C001_R.head(10))
print(C001_R.shape)
```

```python
def keep_highest_confidence(dataset):
    # Ordenar por pupil_timestamp y confidence, y luego eliminar duplicados manteniendo el de mayor confianza
    dataset_sorted = dataset.sort_values(by=['pupil_timestamp', 'confidence'], ascending=[True, False])
    dataset_unique = dataset_sorted.drop_duplicates(subset='pupil_timestamp', keep='first')
    return dataset_unique
```

```python
def plot(dataset, max_threshold, min_threshold, confidence):
    """
    Filtra el dataset por confianza y grafica el diámetro del iris en función del timestamp.
    
    Parameters:
    - dataset: DataFrame con los datos del ojo.
    - max_threshold: Umbral máximo para filtrar los valores del diámetro.
    - min_threshold: Umbral mínimo para filtrar los valores del diámetro.
    """
    print('Tamaño original: ', dataset.shape)
    
    # Filtrar los datos por confianza
    dataset = dataset[dataset['confidence'] > confidence]
    print('Tamaño luego de filtrar por confianza: ', dataset.shape)

    # Filtrar los datos por los umbrales del diámetro
    dataset = dataset[(dataset['diameter'] < max_threshold) & (dataset['diameter'] > min_threshold)]
    
    print('Tamaño luego de filtrar por umbrales: ', dataset.shape)

    # Graficar el diámetro del iris a través del timestamp para el ojo izquierdo
    plt.figure(figsize=(20, 6))
    plt.plot(dataset['pupil_timestamp'], dataset['diameter'], label='dilatación', color='blue')
    plt.xlabel('Timestamp')
    plt.ylabel('Diámetro del iris')
    plt.title('Diámetro del iris a través del tiempo')
    plt.legend()
    plt.grid(True)
    plt.show()
```

### Cálculo de tiempo total y muestras por segundo

```python
def read_export_info(file_path):
    # Leer el archivo export_info.csv
    export_info = pd.read_csv(file_path)
    
    # Extraer el rango de tiempo relativo (formato: MM:SS.mmm - MM:SS.mmm)
    relative_time_range_str = export_info.loc[export_info['key'] == 'Relative Time Range', 'value'].values[0]
    
    # Extraer el tiempo final del rango relativo
    end_time_str = relative_time_range_str.split(' - ')[1]
    
    # Convertir el tiempo final a segundos
    m, s = map(float, end_time_str.split(':'))
    total_time_seconds = m * 60 + s
    
    return total_time_seconds

def calculate_sampling_rate(total_time_seconds, num_samples):
    # Calcular la tasa de muestreo
    sampling_rate = num_samples / total_time_seconds
    return sampling_rate
```

```python
# Leer el archivo export_info.csv para obtener la duración total del experimento
file_path_export_info = 'Data/DataPupila/C001/export_info.csv'
total_time_seconds = read_export_info(file_path_export_info)

# Extraer la cantidad de muestras del ojo izquierdo
num_samples = C001_L.shape[0]

# Calcular la tasa de muestreo
sampling_rate = calculate_sampling_rate(total_time_seconds, num_samples)
print(f'Tasa de muestreo: {sampling_rate:.2f} muestras por segundo')
```

```python
def calculate_sampling_rates(data_frame, patient_code):
    # Generar la ruta del archivo export_info.csv basado en el código del paciente
    export_info_path = f'Data/DataPupila/{patient_code}/export_info.csv'
    
    # Leer la información de exportación
    total_time_seconds = read_export_info(export_info_path)

    # Dividir los datos por ojo
    left_eye_data, right_eye_data = split_eye_data(data_frame)

    # Filtrar la repetición de Timestamps
    left_eye_data = keep_highest_confidence(left_eye_data)
    right_eye_data = keep_highest_confidence(right_eye_data)
    
    # Calcular la cantidad de muestras para cada ojo
    num_samples_left = left_eye_data.shape[0]
    num_samples_right = right_eye_data.shape[0]
    
    # Calcular la tasa de muestreo para cada ojo
    sampling_rate_left = calculate_sampling_rate(total_time_seconds, num_samples_left)
    sampling_rate_right = calculate_sampling_rate(total_time_seconds, num_samples_right)
    
    return sampling_rate_left, sampling_rate_right
```

```python
import plotly.graph_objects as go

def plot_pupil_diameter_plotly(dataset, max_threshold, min_threshold, confidence):
    """
    Filtra el dataset por confianza y grafica el diámetro del iris en función del timestamp usando plotly.
    
    Parameters:
    - dataset: DataFrame con los datos del ojo.
    - max_threshold: Umbral máximo para filtrar los valores del diámetro.
    - min_threshold: Umbral mínimo para filtrar los valores del diámetro.
    - confidence: Umbral mínimo para filtrar los valores de confianza.
    """

    # Dividir los datos por ojo
    L_db, R_db = split_eye_data(dataset)
    L_db = keep_highest_confidence(L_db)
    R_db = keep_highest_confidence(R_db)

    # Filtrar los datos por confianza
    L_db = L_db[L_db['confidence'] > confidence]
    R_db = R_db[R_db['confidence'] > confidence]

    # Filtrar los datos por los umbrales del diámetro
    L_db = L_db[(L_db['diameter'] < max_threshold) & (L_db['diameter'] > min_threshold)]
    R_db = R_db[(R_db['diameter'] < max_threshold) & (R_db['diameter'] > min_threshold)]
    
    # Crear la figura para el ojo izquierdo
    fig_left = go.Figure()
    fig_left.add_trace(go.Scatter(x=L_db['pupil_timestamp'], y=L_db['diameter'],
                                  mode='lines', name='Ojo izquierdo', line=dict(color='blue')))
    fig_left.update_layout(title='Diámetro del iris a través del tiempo (Ojo izquierdo)',
                           xaxis_title='Timestamp',
                           yaxis_title='Diámetro del iris',
                           template='plotly_white')

    # Mostrar la figura del ojo izquierdo
    fig_left.show()

    # Crear la figura para el ojo derecho
    fig_right = go.Figure()
    fig_right.add_trace(go.Scatter(x=R_db['pupil_timestamp'], y=R_db['diameter'],
                                   mode='lines', name='Ojo derecho', line=dict(color='red')))
    fig_right.update_layout(title='Diámetro del iris a través del tiempo (Ojo derecho)',
                            xaxis_title='Timestamp',
                            yaxis_title='Diámetro del iris',
                            template='plotly_white')

    # Mostrar la figura del ojo derecho
    fig_right.show()
```

```python
def read_stimuli_order(patient_code):
    # Generar la ruta del archivo txt basado en el código del paciente
    stimuli_file_path = f'Data/StimuliCode/{patient_code}.txt'
    
    # Leer el archivo
    with open(stimuli_file_path, 'r') as file:
        lines = file.readlines()
    
    # Extraer el orden de los estímulos y su tipo
    stimuli_order = []
    for line in lines[2:]:  # Ignorar las primeras dos líneas (código del paciente y fecha)
        _, stimulus = line.strip().split('\t')  # Separar el índice del estímulo
        stimuli_order.append(stimulus)
    
    return stimuli_order
```

```python
import pandas as pd
import plotly.graph_objects as go

def plot_stimuli_responses(patient_code, start_index, eye_side='l'):
    # Leer el archivo stimuli order para obtener el orden de los estímulos
    stimuli_order = read_stimuli_order(patient_code)
    
    # Leer la base de datos pupil_positions.csv
    data_frame = pd.read_csv(f'Data/DataPupila/{patient_code}/pupil_positions.csv')
    
    # Dividir los datos por ojo y filtrar la repetición de timestamps
    left_eye_data, right_eye_data = split_eye_data(data_frame)
    
    if eye_side == 'l':
        eye_data = keep_highest_confidence(left_eye_data)
        eye_label = 'Ojo izquierdo'
    elif eye_side == 'r':
        eye_data = keep_highest_confidence(right_eye_data)
        eye_label = 'Ojo derecho'
    else:
        raise ValueError("El parámetro 'eye_side' debe ser 'l' (left) o 'r' (right).")

    # Reiniciar los índices después de filtrar
    eye_data = eye_data.reset_index(drop=True)
    
    # Calcular la tasa de muestreo
    sampling_rate_left, sampling_rate_right = calculate_sampling_rates(patient_code)
    
    # Usar la tasa de muestreo adecuada para el ojo seleccionado
    sampling_rate = sampling_rate_left if eye_side == 'l' else sampling_rate_right
    
    # Convertir tiempos en segundos a muestras
    ramp_duration_samples = int(60 * sampling_rate)  # 1 minuto de rampa
    flash_duration_samples = int(1 * sampling_rate)  # 1 segundo de flash
    dark_duration_samples = int(60 * sampling_rate)  # 1 minuto de oscuridad

    # Inicializar la lista de figuras para plotly
    figures = []

    current_index = start_index
    
    for i, stimulus in enumerate(stimuli_order):
        if 'ramp' in stimulus:
            stimulus_duration = ramp_duration_samples
        elif 'flash' in stimulus:
            stimulus_duration = flash_duration_samples
        else:
            raise ValueError(f"Tipo de estímulo desconocido: {stimulus}")
        
        # Calcular el rango de índices para este estímulo
        end_index = current_index + stimulus_duration + dark_duration_samples
        print('Initial sample: ', current_index)
        print('End sample: ', end_index)
        
        if i == len(stimuli_order) - 1:
            # Si es el último estímulo, graficar hasta el final de los datos
            segment_data = eye_data.iloc[current_index:]
        else:
            segment_data = eye_data.iloc[current_index:end_index]
        
        # Calcular el tiempo en segundos desde el inicio del primer estímulo
        segment_time = (segment_data.index - start_index) / sampling_rate
        
        # Crear la figura para el estímulo, usando el número de muestra en el eje x
        fig = go.Figure()
        fig.add_trace(go.Scatter(
            x=segment_data.index, 
            y=segment_data['diameter'],
            mode='lines', 
            name=stimulus,
            hovertemplate='Muestra: %{x}<br>Tiempo: %{customdata:.2f} s<br>Diámetro: %{y:.2f}',
            customdata=segment_time  # Pasar el tiempo en segundos para mostrar en el hover
        ))
        fig.update_layout(title=f'Respuesta pupilar a {stimulus} ({eye_label})',
                          xaxis_title='Número de muestra',
                          yaxis_title='Diámetro del iris',
                          template='plotly_white')
        figures.append(fig)

        # Actualizar el índice actual para el siguiente estímulo
        current_index = end_index
    
    # Mostrar todas las figuras
    for fig in figures:
        fig.show()
```

```python
def find_closest_index(patient_code, source_eye, target_eye, source_index):
    # Leer la base de datos pupil_positions.csv
    data_frame = pd.read_csv(f'Data/DataPupila/{patient_code}/pupil_positions.csv')
    
    # Dividir los datos por ojo
    left_eye_data, right_eye_data = split_eye_data(data_frame)
    
    # Calcular las tasas de muestreo para ambos ojos
    sampling_rate_left, sampling_rate_right = calculate_sampling_rates(patient_code)

    left_eye_data = keep_highest_confidence(left_eye_data)
    right_eye_data = keep_highest_confidence(right_eye_data)

    left_eye_data = left_eye_data.reset_index(drop=True)
    right_eye_data = right_eye_data.reset_index(drop=True)
    
    # Seleccionar la tasa de muestreo y los datos correspondientes a los ojos de origen y destino
    if source_eye == 'l':
        source_data = left_eye_data
        target_data = right_eye_data
        source_sampling_rate = sampling_rate_left
        target_sampling_rate = sampling_rate_right
    elif source_eye == 'r':
        source_data = right_eye_data
        target_data = left_eye_data
        source_sampling_rate = sampling_rate_right
        target_sampling_rate = sampling_rate_left
    else:
        raise ValueError("El parámetro 'source_eye' debe ser 'l' (left) o 'r' (right).")
    
    # Calcular el tiempo correspondiente al índice dado en el ojo de origen
    source_time = source_index / source_sampling_rate
    
    # Calcular el índice correspondiente en el ojo de destino
    target_index = int(round(source_time * target_sampling_rate))
    
    # Asegurarse de que el índice calculado esté dentro de los límites del ojo de destino
    if target_index < 0:
        target_index = 0
    elif target_index >= len(target_data):
        target_index = len(target_data) - 1
    
    return target_index
```

```python
def plot_stimuli_responses(patient_code, start_timestamp, eye_side='l'):
    # Leer el archivo stimuli order para obtener el orden de los estímulos
    stimuli_order = read_stimuli_order(patient_code)
    
    # Leer la base de datos pupil_positions.csv
    data_frame = pd.read_csv(f'Data/DataPupila/{patient_code}/pupil_positions.csv')
    
    # Dividir los datos por ojo y filtrar la repetición de timestamps
    left_eye_data, right_eye_data = split_eye_data(data_frame)
    
    if eye_side == 'l':
        eye_data = keep_highest_confidence(left_eye_data)
        eye_label = 'Ojo izquierdo'
    elif eye_side == 'r':
        eye_data = keep_highest_confidence(right_eye_data)
        eye_label = 'Ojo derecho'
    else:
        raise ValueError("El parámetro 'eye_side' debe ser 'l' (left) o 'r' (right).")

    # Reiniciar los índices después de filtrar
    eye_data = eye_data.reset_index(drop=True)
    
    # Calcular la tasa de muestreo
    sampling_rate_left, sampling_rate_right = calculate_sampling_rates(patient_code)
    
    # Usar la tasa de muestreo adecuada para el ojo seleccionado
    sampling_rate = sampling_rate_left if eye_side == 'l' else sampling_rate_right
    
    # Convertir tiempos en segundos a duraciones en timestamps
    ramp_duration = 60  # 1 minuto de rampa en segundos
    flash_duration = 1   # 1 segundo de flash
    dark_duration = 60   # 1 minuto de oscuridad

    # Inicializar la lista de figuras para plotly
    figures = []

    current_timestamp = start_timestamp
    
    for i, stimulus in enumerate(stimuli_order):
        if 'ramp' in stimulus:
            stimulus_duration = ramp_duration
        elif 'flash' in stimulus:
            stimulus_duration = flash_duration
        else:
            raise ValueError(f"Tipo de estímulo desconocido: {stimulus}")
        
        # Calcular el rango de timestamps para este estímulo
        end_timestamp = current_timestamp + stimulus_duration + dark_duration
        print('Initial timestamp: ', current_timestamp)
        print('End timestamp: ', end_timestamp)
        
        # Seleccionar los datos en el rango de timestamps
        segment_data = eye_data[(eye_data['pupil_timestamp'] >= current_timestamp) &
                                (eye_data['pupil_timestamp'] < end_timestamp)]
        
        # Calcular el tiempo en segundos desde el inicio del primer estímulo
        segment_time = (segment_data['pupil_timestamp'] - start_timestamp).values
        
        # Crear la figura para el estímulo, usando los timestamps en el eje x
        fig = go.Figure()
        fig.add_trace(go.Scatter(
            x=segment_data['pupil_timestamp'], 
            y=segment_data['diameter'],
            mode='lines', 
            name=stimulus,
            hovertemplate=(
                'Muestra: %{customdata[0]}<br>'
                'Timestamp: %{x:.6f}<br>'
                'Tiempo desde inicio: %{customdata[1]:.2f} s<br>'
                'Diámetro: %{y:.2f}'
            ),
            customdata=np.stack((segment_data.index, segment_time), axis=-1)  # Pasar el índice y el tiempo desde inicio para mostrar en el hover
        ))
        fig.update_layout(title=f'Respuesta pupilar a {stimulus} ({eye_label})',
                          xaxis_title='Timestamp',
                          yaxis_title='Diámetro del iris',
                          template='plotly_white')
        figures.append(fig)

        # Actualizar el timestamp actual para el siguiente estímulo
        current_timestamp = end_timestamp
    
    # Mostrar todas las figuras
    for fig in figures:
        fig.show()
```
